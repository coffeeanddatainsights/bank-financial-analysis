```python
import sys
sys.path.append('../src')   # aggiunge la cartella src al path
from utils import *         # importa tutto da utils

df = pd.read_csv("../data/processed/dataset_clean.csv")
```

```python
# 0. Overview
check_df_info(df)

# Gestione Divisioni per 0
df = df.replace([np.inf, -np.inf], np.nan)
df = df.interpolate().fillna(0)
```

```python
# Redditività
df['roa'] = df['net_income']/df['total_assets']
df['roe'] = df['net_income']/df['shareholder_equity']
```

```python
# Efficienza
df['cost_income_ratio'] = df['operating_expenses']/df['operating_income']

# Margine Di Interesse
df['net_interest_margin'] = (df['interest_income'] - df['interest_expense']) / df['average_earning_assets']
```

```python
# Tasso di Crescita % rispetto al gg precedente (YoY)
df['asset_growth'] = df['total_assets'].pct_change()
df['income_growth'] = df['net_income'].pct_change()
```

```python
# Salvo in nuovo dataset
df.to_csv('../data/processed/dataset_transformed.csv', index=False)
```
